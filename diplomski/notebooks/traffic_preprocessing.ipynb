```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from datetime import date
import holidays
```

```python
traffic_data=(r'..\data\traffic_data.xlsx')

df_ulice_podaci = pd.read_excel(traffic_data, engine='openpyxl', sheet_name='traffic_data', index_col=0)


df_ulice_podaci=df_ulice_podaci['2018-01-01':'2020-03-15']

df_ulice_podaci.index = pd.to_datetime(df_ulice_podaci.index)
```

```python
df_ulice_podaci['Days']= pd.to_datetime(df_ulice_podaci.index)

df_ulice_podaci["DayOfWeek"] = df_ulice_podaci['Days'].dt.weekday

a = df_ulice_podaci.DayOfWeek.replace([0,1,2,3,4],'Radni_dan')


a = a.replace([5,6],'Vikend')

df_ulice_podaci = df_ulice_podaci.drop("DayOfWeek",axis=1)

df_ulice_podaci_new = pd.concat([a,df_ulice_podaci],axis = 1, join='inner')

df_ulice_podaci_new = df_ulice_podaci_new.drop('Days',axis=1)

df_ulice_podaci_new
```
