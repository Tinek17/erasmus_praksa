```python
def weather_fun(weather,title,y_os):
    df_fun = data.copy()
    Year = df_fun.index
    x = df_fun[weather]
    plt.figure(figsize=(20,8))
    plt.xticks(range(0,805,30),rotation=90)
    plt.plot(Year, x)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel(y_os)
    plt.show()
```

```python
def kelvinToCelsius(kelvin):
    return kelvin - 273.15
```

```python
T = kelvinToCelsius(data.Temperature_Air_2m_Mean_24h)
data.replace(to_replace = data.Temperature_Air_2m_Mean_24h.values, value = T, inplace= True)
```

```python
T1 = kelvinToCelsius(data.Dew_Point_Temperature_2m_Mean)
data.replace(to_replace = data.Dew_Point_Temperature_2m_Mean.values, value = T1, inplace= True)
```

```python
titles = ['Temperatura', 'Naoblaka', 'Vjetar', 'Rosište', 'Tlak pare']
```

```python
y_os = ['Temperatura, °C', 'Naoblaka, %','Vjetar, m/s','Rosište, °C', 'Tlak pare, %']
```

```python
for i,t,y in zip(data.columns,titles,y_os):
    graf = weather_fun(i,t,y)
```

```python
data = data.fillna(method='ffill')
```
