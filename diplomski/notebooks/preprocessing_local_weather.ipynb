```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
from matplotlib.pyplot import figure
```

```python
data_path = r'../data/full_data.csv'
```

```python
data = pd.read_csv(data_path,index_col=0)
```

```python
data.columns
```

```python
PM = data.filter(like='PM10K',axis=1)
```

```python
PM.index.name = 'index'
```

```python
PM.reset_index(inplace=True)
```

```python
PM['index'] = pd.to_datetime(PM['index'])
```

```python
PM = PM.groupby(pd.Grouper(key='index', freq='D')).mean()
```

```python
PM = PM['2018-01-01':'2020-03-15']
```

```python
PM
```

```python
PM.index = pd.to_datetime(PM.index)
```

```python
PM.to_csv('../data/PM.csv')
```
