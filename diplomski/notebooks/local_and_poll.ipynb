```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
PM = pd.read_csv(r'../data/PM.csv',index_col=0)
```

```python
local = pd.read_csv(r'../data/local_weather.csv',index_col=0)
```

```python
sites = ['Don Bosco','Sjever','Istok','Jug','Zapad',]
```

```python
def pollutanti(site_1,pollution):
    df_fun = PM.copy()
    Year = df_fun.index
    x = df_fun[pollution]
    plt.figure(figsize=(20,8))
    plt.xticks(range(0,805,30),rotation=90)
    plt.plot(Year, x)
    plt.title(site_1)
    plt.xlabel('Date')
    plt.ylabel('Lebdeće čestice, µm/m³')
    plt.show()
    
```

```python
for site,i in zip(sites,PM.columns):
    graf = pollutanti(site,i)
```
