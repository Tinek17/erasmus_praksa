# Read air pollution data from 2010-01-01

```python
path = '../../data/Daily_historical_data/O3,NO2,PM10-used_in_paper2020'
df2010 = import_data_year2010(path)
```

```python
df2010
```

## Impute missing values with bfill

```python
df2010.isnull().sum()/df2010.shape[0]*100
```

```python
df2010_nonan = df2010.fillna(method='bfill')
df2010_nonan.isnull().sum().any()
```

## Add pollution measure total oxidant, Ox
Ox = NO2 + O3

```python
df = df2010_nonan.copy()
df['N_Ox'] = df['N_NO2'] + df['N_O3']
df['S_Ox'] = df['S_NO2'] + df['S_O3']
```

## Add temporal features

```python
df_temp = temporal_feat(df)
df_temp.head(2)
```

```python
df_temp[['dayofyear']]
```

### Temporal dummies for season, month and weekday

```python
temp_cols = ['month', 'weekday', 'season']
df_temp = temporal_dummies(df_temp, temp_cols)
df_temp.head(2)
```

### Export (a)

```python
#df_temp.to_csv('../../results/5sta(DNOWS)_3com(NO2,PM10K,O3)_year2010_TemporalDummies_Ox.csv')
```

## Add public holidays feature

```python
df_temp.index[0]#.split()[0]
```

```python
df_temp = public_holidays(df_temp)
df_temp.head(2)
```

## Add school holidays feature

# Add meteorological data

```python
path = '../../data/Daily_historical_data/RH,Temp,Precip,Press,Wind_speed&direction-used_in_paper2020'
df_meteo = import_meteo_year2010(path)
```

```python
df_meteo
```

## Impute missing values with bfill, except for Ost station
* **In Ost station data starts from 2017-05-18**

```python
df_meteo.isnull().sum()/df_meteo.shape[0]*100
```

```python
nonan = df_meteo.drop(['Ost_Pressure','Ost_RH','Ost_Temp','Ost_Winddirection','Ost_Windspeed'], axis=1).fillna(method='bfill')
df_meteo_new = pd.concat([nonan, df_meteo[['Ost_Pressure','Ost_RH','Ost_Temp','Ost_Winddirection','Ost_Windspeed']]], axis=1)
```

# Join air pollution and meteorological data

```python
df_all = df_temp.join(df_meteo_new)
df_all
```
