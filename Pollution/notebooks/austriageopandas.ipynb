```python
import folium 
```

```python
m = folium.Map(location = [47.5162, 14.5501], zoom_start = 7,tiles='cartodbpositron')
```

```python
geo=r'..\data\map.geojson'
styria=r'..\data\styria.geojson'
```

```python
bordersStyle={
    'color':'blue',  
    'fillOpacity':0.05
}
```

```python
folium.GeoJson(geo,name='Austrija',style_function=lambda x:bordersStyle).add_to(m)
```

```python
bordersStyle2={
    'color':'darkblue', 
    'fillOpacity':0.5
}
```

```python
folium.GeoJson(styria,name='Styria',style_function=lambda x:bordersStyle2).add_to(m)
```

```python
folium.Marker(location=[47.0707,15.4395],popup='Graz',icon=folium.Icon(icon="fa-solid fa-city",color='red')).add_to(m)
```
