# Pretvaranje iz K u °C

```python
def kelvinToCelsius(kelvin):
    return kelvin - 273.15
```

```python
T = kelvinToCelsius(df.Temperature_Air_2m_Mean_24h)
df.replace(to_replace = df.Temperature_Air_2m_Mean_24h.values, value = T, inplace= True)
```

```python
T1 = kelvinToCelsius(df.Dew_Point_Temperature_2m_Mean)
df.replace(to_replace = df.Dew_Point_Temperature_2m_Mean.values, value = T1, inplace= True)
```

# Za grafove

```python
factors = ['Cloud_Cover_Mean', 'Dew_Point_Temperature_2m_Mean',
       'Temperature_Air_2m_Mean_24h', 'Vapour_Pressure_Mean',
       'Wind_Speed_10m_Mean']
df.index = pd.to_datetime(df.index)
```

# Heatmap

```python
for factor in factors:
        df_f = df[[factor]].loc['2019']
        df_f['month'] = df_f.index.month
        df_f['day'] = df_f.index.day
        df_pivot = df_f.pivot(index='day', columns='month', values=factor)

        plt.figure(figsize=(8, 6), dpi=80)
        ax = sns.heatmap(df_pivot,linewidths=.5,cmap="YlGnBu")
        #plt.title(year+'\n'+factor, fontdict={'fontsize':15})
        plt.title('2019'+', '+factor, fontdict={'fontsize':15})
        plt.xlabel('Months', fontsize = 15) # x-axis label with fontsize 15
        plt.ylabel('Days', fontsize = 15) # y-axis label with fontsize 15
        #plt.yticks(rotation=90)
```

# Linijski

```python
for factor in factors:
    Year = df.index
    x = df[factor]
    plt.figure(figsize=(20,8))
    plt.plot(Year, x,)
    plt.title(factor)
    plt.xlabel('Date')
    plt.ylabel(factor)
    plt.show()
    day = df[factor].idxmax()
    max_factor = round(df[factor].max(),2)
    print(f'Maksimalna vrijednost {max_factor} i datum {day}')
```

# Histogram

```python
for factor in factors:
    Year = df.index
    x = df[factor]
    plt.figure(figsize=(20,8))
    plt.hist(x, bins = 50, color = "skyblue")
    plt.title(factor)
    plt.xlabel(factor)
    plt.ylabel('Times')
    plt.show()
```

# Oblaci 

```python
clear_sky=0
scattered_cloluds=0
broken_clouds=0
Overcast = 0
for i in df.Cloud_Cover_Mean:
    if i <= 0.25:
        clear_sky =clear_sky+ 1
    elif i <= 0.75:    
        scattered_cloluds=scattered_cloluds+1
    elif i < 100:
        broken_clouds =broken_clouds+1
    elif i == 100:
        Overcast =Overcast+1
print (f'Clear Skies {clear_sky}')
print (f'Scattered Cloluds {scattered_cloluds}')    
print (f'Broken Clouds {broken_clouds}')    
print (f'Overcast {Overcast}')
```
