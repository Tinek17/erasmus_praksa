```python
# sys packages
import os,sys
sys.path.append('..')
from math import pi
# 3rd party packages
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import numpy as np
from windrose import WindroseAxes
from IPython.display import Image
from IPython.core.display import HTML
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
import dateutil.parser
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams["axes.titleweight"] = "bold"
import itertools
```

```python
df=pd.read_csv('..\Windrose Pollution Know - Center\\5sta(DNOWS)_3com(NO2,PM10K,O3)_year2010_TemporalDummies_Ox_holidays_meteo.csv', index_col=0)
df.index = pd.to_datetime(df.index)

df_1 = df.dropna(axis = 0, how ='any')
```

```python
factors_wind_speed = ['Nord_Windspeed','Sud_Windspeed','West_Windspeed','Ost_Windspeed']
factors_wind_direction = ['Nord_Winddirection','Sud_Winddirection','West_Winddirection','Ost_Winddirection']
```

```python
for factor_ws in factors_wind_speed:
    for factor_wd in factors_wind_direction:
        df_windrose_factor_ws = df_1[factor_ws]
        df_windrose_factor_wd = df_1[factor_wd]
        ax = WindroseAxes.from_ax()
        ax.bar(df_windrose_factor_wd, df_windrose_factor_ws,opening=.9,edgecolor='white')
        ax.set_title(factor_wd)
        ax.legend(bbox_to_anchor=(1.0, 1.0),fontsize=16)
        ax.set_xticklabels(['N', 'NW',  'W', 'SW', 'S', 'SE','E', 'NE'])
        ax.set_theta_zero_location('N')
        plt.rc('font', family='Times New Roman', weight='bold', size=20)
```
