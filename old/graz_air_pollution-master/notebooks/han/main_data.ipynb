```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/01_working_data/main_data.csv', index_col=0)
df
```

```python
des = df.describe().T[['count', 'mean', 'min', 'max']]
des
```

```python
des['place'] = des.index.str.split().str[0]
des['param'] = des.index.str.split('(').str[-1].str[:-1]
des
```

```python
for p in sorted(des['param'].unique()):
    tmp = des[des['param']==p].drop(['count', 'param'], axis=1).set_index('place').unstack().reset_index()
    tmp[0] = tmp[0].round(1)
    fig = px.bar(tmp, color='place', y=0, x='level_0', barmode='group',
           width=600, height=350, text=0,
         title=f'<b>{p}</b>'
          )
    fig.update_layout(margin=dict({'t':40}))
    fig.show()
#     break
```

```python
places = des['place'].unique()
params = des['param'].unique()
places, params
```

# Aggregations

## Yearly

```python
df.reset_index(inplace=True)
```

```python
df['index'] = pd.to_datetime(df['index'])
df['year'] = df['index'].dt.year
```

```python
a = df.groupby('year').count().T.iloc[1:, :]
a
```

```python
a[a<341]
```

```python
a = df.groupby(pd.Grouper(key='index', freq='Y')).mean()
for p in params:
    cols = [i for i in a.columns if f'({p})' in i]
    tmp = a[cols]
    fig = px.line(tmp, width=650, height=350, title=f'<b>Yearly Average: {p}</b>')
    fig.update_layout(margin=dict({'t':40}))
    fig.show()
#     break
```

## Monthly

```python
a = df.groupby(pd.Grouper(key='index', freq='m')).mean()
for p in params:
    cols = [i for i in a.columns if f'({p})' in i]
    tmp = a[cols]
    fig = px.line(tmp, width=650, height=350, title=f'<b>Monthly Average: {p}</b>')
    fig.update_layout(margin=dict({'t':40}))
    fig.show()
#     break
```
