```python
import pandas as pd 
import numpy as np 
```

```python
data_path = ('..\data\working_data.xlsx')
```

```python
xls = pd.ExcelFile(data_path)
print(xls.sheet_names)
```

```python
list_of_df=[]
for sheet in xls.sheet_names:
    print(sheet)
    xf = pd.read_excel(data_path,sheet_name=sheet,index_col=0)
    list_of_df.append(xf)
```

```python
full_df = pd.concat(list_of_df,axis=1)
```

```python
full_df.to_csv('full_data.csv')
```
